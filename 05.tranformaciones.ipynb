```python
U,D,V = np.linalg.svd(A)
```

```python
fig, ax = plt.subplots(2,2, figsize=(10,10))
fig.suptitle("Transformaciones")
fig.patch.set_facecolor('moccasin')
ax[0,0].set_title("Sin transformar")
graficarMatriz(ax[0][0], np.identity(2))
ax[0,1].set_title("Primera Rotacin ($V$)")
graficarMatriz(ax[0][1], V)
ax[1,0].set_title("Escala ($D$)")
graficarMatriz(ax[1][0], np.diag(D) @ V)
ax[1,1].set_title("Segunda Rotacin ($U$)")
graficarMatriz(ax[1,1], U @ np.diag(D) @ V )
plt.show()
```
